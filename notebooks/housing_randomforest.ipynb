```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from pathlib import Path
```

```python
INPUT_PATH = Path.cwd() / "data"
```

```python
df = pd.read_csv(f"{INPUT_PATH / 'train.csv'}")
```

```python
for c in df.columns:
    
    if df[c].dtypes == "object":
        df[c] = df[c].astype("category")
        df[c] = df[c].cat.codes
```

```python
df = df.drop(columns=["LotFrontage", "MasVnrArea", "GarageYrBlt"])
```

```python
train_X = df.loc[:int(.8*len(df))-1, df.columns != "SalePrice"]
train_y = np.log(df.loc[:int(.8*len(df))-1, "SalePrice"])

valid_X = df.loc[int(.8*len(df)):, df.columns != "SalePrice"]
valid_y = np.log(df.loc[int(.8*len(df)):, "SalePrice"])
```

```python
train_X.tail()
```

```python
valid_X.head()
```

```python
m = RandomForestRegressor(n_jobs=-1)
```

```python
m.fit(train_X, train_y)
```

```python
import math
def rmse(y, yhat):
    return math.sqrt(((y-yhat)**2).mean())
```

```python
valid_yhat = m.predict(valid_X)
rmse(valid_y, valid_yhat)
```
